# 1. Create a tuple of five countries and print the third country

```python
countries = ("Ghana", "Togo", "Nigeria", "Kenya", "USA")
print(countries[2])
```

# 2. Create a tuple of numbers and print the last number using negative indexing

```python
numbers = (1, 2, 3, 4, 5)
print(numbers[-1])
```

# 3. Slice a tuple to display the first three items

```python
food = ("rice", "beans", "yam", "plantain", "fufu")
print(food[:3])
```

# 4. Count how many times 5 appears in a tuple

```python
numbers = (5, 2, 5, 7, 8, 5, 10)
print(numbers.count(5))
```

# 5. Find the index of "Python" in a tuple of programming languages.

```python
languages = ("Python", "Java", "C++", "JavaScript", "Ruby")
print(languages.index("Python"))
```

# 6. Concatenate two tuples

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = tuple1 + tuple2
print(result)
```

# 7. Repeat a tuple three times

```python
letters = ('a', 'b', 'c', 'd', 'e')
print(letters * 3)
```

# 8. Unpack a tuple containing a student's name, age and grade

```python
student = ("Cypher", 20, "A")
name, age, grade = student
print("Name:", name)
print("Age:", age)
print("Grade:", grade)
```

# 9. Reverse a tuple using slicing

```python
numbers = (1, 2, 3, 4, 5)
print(numbers[::-1])
```

# 10. Convert a tuple to a list, modify one element and convert it back to a tuple

```python
colours = ("red", "green", "blue", "yellow")
colours_list = list(colours)
colours_list[1] = "purple"
colours = tuple(colours_list)
print(colours)
```
