# What is a Set?

A set is an unordered collection of unique items in Python.

A set stores multiple values, but it does not allow duplicate values.

```python
fruits = { "Mango"," Apple","Banana"}

print(fruits)
```

# Characteristics of Sets

1. Unordered

Sets do not store items in a fixed position.

```python
numbers = {1, 2, 3, 4}

x = {3, 1, 4, 2}

print(numbers)
print(x)
```

# You cannot access items using indexes:

```python
numbers[0]
```

# No Duplicate Values

Python keeps only unique values.

```python
numbers = {1, 2, 2, 3, 3, 3}

print(numbers)
```

# Mutable

Sets can be changed after creation.

You can:

Add items

Remove items

Update items

```python
colors = {"Red", "Blue"}

colors.add("Green")

print(colors)
```

```python
colors.remove("Red")
colors
```

```python
colors1 = {"Gold"}
colors.update(colors1)
colors
```

# Cannot Contain Mutable Items

Sets can only contain immutable values.

```python
numbers = {1, 2, 3}
names = {"John", "Mary"}
```

```python
data = {[1,2], [3,4]}
```

# Creating Sets

Correct

Creating a Set with Values

```python
students = {"John", "Mary", "Peter"}

print(students)
```

# Using set()

Convert another collection into a set.

List:

```python
numbers = [1,2,3,3,4]

unique_numbers = set(numbers)

print(unique_numbers)
```

# Checking Items in a Set

Use in.

```python
fruits = {"Apple", "Orange", "Banana"}

print("Apple" in fruits)
```

```python
print("Mango" in fruits)
```

# Adding Items
add()

```python
languages = {"Python", "Java"}

languages.add("JavaScript")

print(languages)
```

# update()

Adds multiple items.

```python
languages = {"Python", "Java"}

languages.update(["C", "C++", "Go"])

print(languages)
```

You can update with another set:

```python
a = {1,2,3}

b = {4,5,6}

a.update(b)

print(a)
```

#  Removing Items
remove()

Removes an item.

```python
colors = {"Red","Blue","Green"}

colors.remove("Blue")

print(colors)
```

# discard()

Similar to remove but does not give an error.

```python
colors.discard("Black")

print(colors)
```

# pop()

Removes a random item.

```python
numbers = {1,2,3,4}

numbers.pop()

print(numbers)
```

# clear()

Removes everything.

```python
numbers = {1,2,3}

numbers.clear()

print(numbers)
```

# Set Length

Use len().

```python
names = {"John","Mary","Peter"}

print(len(names))
```

# Set Operations

This is where sets become powerful.

```python
python_students = {"John","Mary","Peter"}

java_students = {"Mary","David","Peter"}
```

# Union (Combine Sets)

Symbol:    `|` or `union()`

```python
all_students = python_students | java_students

print(all_students)
```

# Intersection (Common Items)

Symbol: `&`

```python
x = python_students & java_students

print(x)
```

# Difference

Items that exist in one set but not another.

Symbol: `-`

```python
only_python = python_students - java_students

print(only_python)
```

# Symmetric Difference

Items that exist in either set but not both.

Symbol: `^`

```python
result = python_students ^ java_students

print(result)
```

## 10. Set Methods Summary

| Method | Purpose |
|---|---|
| `add()` | Add one item to a set |
| `update()` | Add multiple items to a set |
| `remove()` | Remove an item (gives an error if the item is missing) |
| `discard()` | Remove an item safely (no error if the item is missing) |
| `pop()` | Remove and return a random item from the set |
| `clear()` | Remove all items from the set |
| `union()` | Combine two or more sets |
| `intersection()` | Find common items between sets |
| `difference()` | Find items that exist in one set but not another |
| `symmetric_difference()` | Find items that exist in either set but not both |
