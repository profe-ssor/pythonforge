# What is a Dictionary?

A dictionary is a collection of data stored as key-value pairs.

Each item in a dictionary has:

Key → A unique identifier
Value → The data associated with the key
Syntax

# Syntax

```
dictionary_name = {
    key: value
}

````python
student = {
    "name": "John",
    "age": 21,
    "course": "Computer Science"
}

print(student)
````

# Dictionary Characteristics

Since Python 3.7, dictionaries maintain the order in which items are added.

````python
student = {
    "name": "John",
    "age": 21,
    "country": "Ghana"
}

print(student)
````

# Mutable

Dictionaries can be changed after creation.

````python
student = {
    "name": "John",
    "age": 21
}

student["age"] = 22

print(student)
````

# Keys Must Be Unique

A dictionary cannot have duplicate keys.

````python
student = {
    "name": "John",
    "name": "Mary"
}

print(student)
````

# Keys Must Be Immutable

Dictionary keys must be values that cannot change.

````python
student = {
    "name": "John",
    1: "First",
    (1,2): "Coordinate"
}
````

````python
student = {
    [1,2]: "Location"
}
````

# Values Can Be Any Data Type

Dictionary values can store anything.

````python
data = {
    "name": "John",
    "age": 25,
    "skills": ["Python", "Django"],
    "active": True
}
````

# Creating Dictionaries
Empty Dictionary

````python
student = {}

print(student)
````

# Using dict()

````python
student = dict(name="John", age=20)

print(student)
````

# Accessing Dictionary Values
Using Keys

````python
student = {
    "name": "John",
    "age": 21
}

print(student["name"])
````

# Using get()

````python
print(student.get("age"))
````

# Difference Between [] and get()

Using brackets:

````python
student["grade"]
````

````python
student.get("grade")
````

# Adding New Items

````python
student = {
    "name": "John",
    "age": 21
}

student["country"] = "Ghana"

print(student)
````

# Updating Dictionary Items
Change Existing Value

````python
student = {
    "name": "John",
    "age": 21
}

student["age"] = 22

print(student)
````

# Using update()

````python
student.update({
    "age":23,
    "country":"Ghana"
})

print(student)
````

# Removing Dictionary Items
pop()

Removes an item using its key.

````python
student = {
    "name":"John",
    "age":21
}

student.pop("age")

print(student)
````

# popitem()

Removes the last inserted item.

````python
student.popitem()
````

# del

````python
del student["name"]
````

# clear()

Remove everything.

````python
student.clear()

print(student)
````

# Dictionary Length

Use len():

````python
student = {
    "name":"John",
    "age":21,
    "country":"Ghana"
}

print(len(student))
````

## 10. Dictionary Methods Summary

| Method | Purpose |
|---|---|
| `get()` | Access a value safely |
| `keys()` | Return all keys in the dictionary |
| `values()` | Return all values in the dictionary |
| `items()` | Return all key-value pairs |
| `update()` | Add new items or update existing items |
| `pop()` | Remove an item using its key |
| `popitem()` | Remove and return the last inserted item |
| `clear()` | Remove all items from the dictionary |
| `copy()` | Create a copy of the dictionary |
