# What is a Tuple?

A tuple is an ordered collection of items.

Unlike lists, tuples cannot be changed after they are created (they are immutable).

```python
fruits = ("Orange", "Banana", "Apple")

print(fruits)
```

# Tuple Characteristics

1. Ordered
2. Immutable
3. Allows duplicate values
4. Can store different data types
3. Indexed

```python
Student = ("Cypher", 15, True, 3.9)
print(Student)
```

# Creating Tuples

```python
colors = ("Red", "Blue", "Green")
print(colors)
```

# Empty Tuple

```python
empty = ()
print(empty)
```

# Single Item Tuple

```python
fruit = ("Apple",)
print(type(fruit))
```

# Using tuple()

```python
numbers = tuple([1, 2, 3, 4])

print(numbers)
```

# Accessing Tuple Elements

Using indexes.

```python
animals = ("Dog", "Cat", "Lion", "Tiger")

print(animals[0])
print(animals[2])

# Negative Indexing

print(animals[-1])
print(animals[-2])
```

# Tuple Slicing

```python
numbers = (10,20,30,40,50,60)

print(numbers[1:4])

# Reverse 
print(numbers[::-1])
```

# Checking if an Item Exists

```python
fruits = ("Apple", "Orange", "Banana")

print("Orange" in fruits)
print("Mango" in fruits)
```

# Tuple Length len()

```python
names = ("John", "Mary", "Peter")

print(len(names))
```

# Tuple Methods

Tuples only have two built-in methods. 
1. count()
2. indexed()

```python
numbers = (1,2,2,3,2,5)

print(numbers.count(2))
```

```python
numbers = (10,20,30,40)

print(numbers.index(30))
```

# Tuple Operations
Concatenation

```python
a = (1,2,3)
b = (4,5)

print(a + b)
```

# Repetition

```python
print((1,2)*3)
```

# Nested Tuples

```python
students = (
    ("John",20),
    ("Mary",22),
    ("Peter",19)
)

print(students[0])
print(students[1][0])
```

#  Tuple Packing

Python automatically packs values into a tuple.

```python
person = "John", 22, "Ghana"

print(person)
```

# Tuple Unpacking

```python
person = ("John",22,"Ghana")

name, age, country = person

print(name)
print(age)
print(country)
```

# Converting Between Lists and Tuples

# Tuple to List

```python
numbers = (1,2,3)

numbers = list(numbers)

print(numbers)
```

# List to Tuple

```python
numbers = [1,2,3]

numbers = tuple(numbers)

print(numbers)
```

# Why Tuples are Immutable

```python
colors = ("Red","Blue","Green")

colors[0] = "Black"
```

# Modifying a Tuple (Indirectly)

Since tuples cannot be changed directly:

```python
colors = ("Red","Blue","Green")

temp = list(colors)

temp[0] = "Black"

colors = tuple(temp)

print(colors)
```

# Deleting Tuples

Delete the entire tuple.

```python
numbers = (1,2,3)

del numbers
```
