```python
secret_number = 25

attempt = 0

guess_number = int(input("Enter a secret number: "))

while guess_number != secret_number:
    attempt += 1

    if guess_number in range(0, 25):
        
        print(f"{guess_number} is too low")

    elif guess_number in range(26, 100):
        print(f"{guess_number} is too high")

    else:
        print("Enter a number between 0 and 100")

    guess_number = int(input("Enter a secret number: "))

print("Correct! You guessed the number.")
print(f"Your total attempt {attempt}")
```

# Question 6

```python
count_pos = 0
count_neg = 0
stop = 999

number = int(input("Enter an integer number: "))

while number != stop:

    if number < 0:
        count_neg += 1

    elif number > 0:
        count_pos += 1

    number = int(input("Enter an integer number: "))

print("Positive numbers:", count_pos)
print("Negative numbers:", count_neg)
```

# Question 7

```python
stop = -1

counter = 0
total = 0
highest = None
lowest = None

marks = float(input("Enter Student's Mark (-1 to stop): "))

while marks != stop:

    counter += 1
    total += marks

    if highest is None or marks > highest:
        highest = marks

    if lowest is None or marks < lowest:
        lowest = marks

    marks = float(input("Enter Student's Mark (-1 to stop): "))

if counter > 0:
    average = total / counter

    print("\n----- Results -----")
    print(f"Total number of marks entered: {counter}")
    print(f"Average mark: {average:.2f}")
    print(f"Highest mark: {highest}")
    print(f"Lowest mark: {lowest}")
else:
    print("No marks were entered.")
```
